```python
#importing librarires 
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd 
import time 

#navigating Amazon.com
titles=[]
price=[]
rating=[]
c=0
def scraper(keyword,page_no):
    global titles , price , rating , c
    driver = webdriver.Chrome(executable_path=r'C:\Users\aldawlia\Desktop\Demo\chromedriver.exe')
    base='https://www.amazon.com/ref=nav_logo'
    driver.get('https://www.amazon.com/ref=nav_logo')
    print('access successful')
    #overcoming captcha
    print('solve captcha manually and type Enter when finished')
    input('Type Enter :')
    search_box=WebDriverWait(driver,60).until(ec.element_to_be_clickable((By.XPATH,"//input[@id='twotabsearchtextbox']")))
    search_box.click()
    search_box.send_keys(keyword)
    search_box.send_keys(Keys.ENTER)
    time.sleep(10)
    soupy = BeautifulSoup(driver.page_source,'lxml')
    time.sleep(15)
    #selecting brand name
    company_filter=soupy.find('ul', {'data-csa-c-content-id':"2528832011"})
    names = company_filter.find_all('li')
    for name in names:
        if 'aria-label' in name.attrs:
            print(name['aria-label'])
        else :
            continue
            

    choice=input('choose the company please')
    print(choice)
    link = company_filter.find('li', {'aria-label': f"{choice}"}).find('a').get('href')
    print(link)
    time.sleep(30)
    filtered_products = driver.find_element_by_xpath(f"//a[@href='{link}']")
    filtered_products.click()      
    time.sleep(20)
    soup=BeautifulSoup(driver.page_source,'lxml')
    for page in range(page_no):
            divs=soup.select("div.puis-card-container.s-card-container.s-overflow-hidden.aok-relative.puis-include-content-margin.puis.puis-vw0goty2ahaky2ds8ppx5s5mhw.s-latency-cf-section.puis-card-border")
            print(len(divs))
            for div in divs:
                try:
                    title = div.select_one("span.a-size-base-plus.a-color-base.a-text-normal").text
                    if title:
                        titles.append(title)
                    else:
                            raise AttributeError
                except AttributeError:
                    title2=div.select_one("span.a-size-medium.a-color-base.a-text-normal").text
                    titles.append(title2)    
                
                try:    
                    value=div.select_one('span.a-price > span.a-offscreen').text 
                    price.append(value)
                except AttributeError:
                    price.append('misisng')
            
                try :
                    rate=div.select_one('span.a-icon-alt').text[:3]
                    rating.append(rate)
                except AttributeError:
                    rating.append('missing')
                
            time.sleep(20)    
            next_button = driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")    
            if next_button:
                next_button.click()
                print(f'page {c+1} scraped')
                c=c+1
                time.sleep(40)
            else:
                print('end')
                break

    driver.quit()
    df=pd.DataFrame()
    df['product_name']=titles
    df['price']=price
    df['rating']=rating
    df.to_excel('#scraper_project_roma.xlsx',index=False)
scraper('tv',3)                                          
```
